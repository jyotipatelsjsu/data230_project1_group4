```python
list_digit=["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten"]

def fetch_exp(des):
    exp = set()
    desc_text = TextBlob(des)
    sentences = desc_text.sentences
    for sentence in sentences:
        if 'experience' in sentence and 'year' in sentence:
            # print("SENTENCE:", sentence)
            desc_words = TextBlob(str(sentence))
            words = desc_words.words
            # print("WORDS:", words)
            for i in range(0, len(words)):
                if "year" in words[i].lower():
                    exp.add(words[i-1])
    list_exp = list(exp)
    # print(list_exp)
    if len(list_exp) > 0:
        return fetch_experience_digit(list_exp[0])
    else:
        return 0

# TODO If first element in not valid list take second...
def fetch_experience_digit(year_exp):
    if len(year_exp) <= 2:
        _, num = check_if_int(year_exp)
        return num
    else:
        if "-" in year_exp:
            num_split = year_exp.split('-')[0]
            _, num = check_if_int(year_exp)
            return num
        else:
            # TODO get another from list
            if year_exp.lower() in list_digit:
                return list_digit.index(year_exp.lower())
            else:
                return 0
    return 0

def check_if_int(input_num):
    try:
        # converting the inputs to integer and catching the exception
        converted_num = int(input_num)
    except ValueError:
        # printing the error message if the input is not of type int
        # return falseif input is not integer and default value -1
        # print("ERROR: Character is not of type INT in:", input_num)
        return False, 0
    return True, converted_num
```

```python
df = data.copy()
text = df.description.values
exp_col = list()
for t in text:
    exp_list = fetch_exp(str(t))
    # print("TEST:", exp_list)
    exp_col.append(int(exp_list))
print(exp_col)
```

```python
df['experience'] = exp_col
```
