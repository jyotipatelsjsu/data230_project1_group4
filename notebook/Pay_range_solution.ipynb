```python
import pandas as pd
df= pd.read_csv(r'C:\Users\HP\Desktop\Data 230\project\project_230_sruthi.csv')
```

```python
pay_range=[]
```

```python
word1= 'analyst'
word2= 'scientist'
word3= 'machine'
word4= 'engineer'
```

```python
for value in df['position']:
    print(value)
    if value.find(word1)!=-1:
        pay_range.append("$80k-$100k")
    elif value.find(word2)!=-1:
        pay_range.append("$100k-$125k")
    elif value.find(word3)!=-1:
        pay_range.append("$120-$150")
    elif value.find(word4)!=-1:
        pay_range.append("$110-$130")
    else:
        pay_range.append("Pay range varies")

df["Pay_range"]= pay_range
```
